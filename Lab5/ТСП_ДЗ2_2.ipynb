```python
import numpy as np
import pandas as pd
import scipy.stats as sps
import math
import matplotlib.pyplot as plt
```

```python
T = 14
h = 0.1
N = int((T - 0)/h)
n = 100
sigma = 0.25
z = 1.5
N
```

```python
Ksi = np.zeros((n, 2*N, 2))
for i in range(0, n):
    for j in range(0, 2*N):
        Ksi[i][j][0] = sps.norm(0, sigma * np.sqrt(h/2)).rvs(size = 1)
        Ksi[i][j][1] = sps.norm(0, sigma * np.sqrt(h/2)).rvs(size = 1)

#с шагом h
trajectory_h = np.zeros((n, N + 1, 2))
for i in range(0, n):
    for j in range(1, N+1):
        trajectory_h[i][j][0] = trajectory_h[i][j - 1][0] + Ksi[i][2*j - 2][0] + Ksi[i][2*j - 1][0]
        trajectory_h[i][j][1] = trajectory_h[i][j - 1][1] + Ksi[i][2*j - 2][1] + Ksi[i][2*j - 1][1]

#с шагом h/2
trajectory_h2 = np.zeros((n, 2*N + 1, 2))
for i in range(0, n):
    for j in range(1, 2*N+1):
        trajectory_h2[i][j][0] = trajectory_h2[i][j - 1][0] + Ksi[i][j - 1][0]
        trajectory_h2[i][j][1] = trajectory_h2[i][j - 1][1] + Ksi[i][j - 1][1]
```

```python
trajectory_hx = np.zeros((N+1))
trajectory_hy = np.zeros((N+1))
for j in range(N+1):
  trajectory_hx[j] = trajectory_h[0][j][0]
  trajectory_hy[j] = trajectory_h[0][j][1]

trajectory_h2x = np.zeros((2*N+1))
trajectory_h2y = np.zeros((2*N+1))
for j in range(2*N+1):
  trajectory_h2x[j] = trajectory_h2[0][j][0]
  trajectory_h2y[j] = trajectory_h2[0][j][1]

plt.figure(figsize=(12, 8))
plt.plot(trajectory_hx, trajectory_hy, linestyle='-', marker='o', markersize = 3, linewidth=1, color='orange', label="h")
plt.plot(trajectory_h2x, trajectory_h2y, linestyle='--', marker='o', markersize = 3, linewidth=1, color='crimson', label="h/2")
plt.grid()
plt.legend()
plt.show()
```

```python
sum_h = np.zeros(2)
sqSum_h = np.zeros(2);

for i in range(0, n):
    for j in range(1, N+1):
      sum_h[0] +=abs(trajectory_h[i][j][0] -  trajectory_h[i][j-1][0])
      sum_h[1] +=abs(trajectory_h[i][j][1] -  trajectory_h[i][j-1][1])
      sqSum_h[0] +=abs(trajectory_h[i][j][0] -  trajectory_h[i][j-1][0])** 2
      sqSum_h[1] +=abs(trajectory_h[i][j][1] -  trajectory_h[i][j-1][1])** 2

Var_h = np.zeros(2);
SqVar_h = np.zeros(2);

Var_h[0] = sum_h[0]/n
Var_h[1] = sum_h[1]/n
SqVar_h[0] = sqSum_h[0]/n
SqVar_h[1] = sqSum_h[1]/n

((sum_h[0], sum_h[1]), (Var_h[0], Var_h[1]), (sqSum_h[0], sqSum_h[1]), (SqVar_h[0], SqVar_h[1]))
```
