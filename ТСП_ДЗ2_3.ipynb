```python
import numpy as np
from sympy import *

s = Symbol('s')
alpha = Symbol('alpha')

#Запишем матрицу коэффициентов при (p_1 ) ̃, (p_2 ) ̃, (p_3 ) ̃, (p_4 ) ̃, (p_5 ) ̃, (p_6 ) ̃, (p_7 ) ̃:
M = Matrix([[s+3, -2, 0, 0, 0, 0, 0], [-1, s+5, 0, 0, 0, 0, 0], [0, 0, s+4, -5, 0, 0, -6], 
                  [0, -3, 0, s+5, 0, 0, -3], [-1, 0, 0, 0, s+5, -2, 0], [0, 0, 0, 0, -5, s+2, 0], [-1, 0, -4, 0, 0, 0, s+9]])

#Также запишем вектор свободных коэффициентов:
Vec = Matrix([alpha, 1-alpha, 0, 0, 0, 0, 0])

#Решим матричное уравнение:
Sol = simplify(M**(-1)*Vec)
print(Sol)
```

```python
cancel(s*(s**4 + 26*s**3 + 234*s**2 + 850*s + 1001))
```

```python
#Запишем матрицу коэффициентов при (p_1 ) ̃, (p_2 ) ̃:
M1 = Matrix([[s+3, -2], [-1, s+5]])

#Также запишем вектор свободных коэффициентов:
Vec1 = Matrix([alpha, 1-alpha])

#Решим матричное уравнение:
Sol1 = simplify(M1**(-1)*Vec1)
Sol1
```

```python
print(Sol1)
```

```python
t = Symbol('t', positive = True)
p_1 = together(simplify(powsimp(expand(inverse_laplace_transform(Sol1[0], s, t)))))
p_2 = together(simplify(powsimp(expand(inverse_laplace_transform(Sol1[1], s, t)))))

p = Matrix([p_1, p_2])
p
```
